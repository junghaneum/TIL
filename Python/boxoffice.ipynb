```python
date=[]
dt_index = pd.date_range(start='20190617', end='20200531')
dt_list = dt_index.strftime("%Y%m%d").tolist()
for i in dt_list:
    date.append(i)
```

```python
data=[]
for i in range(50):
    url =f"http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchWeeklyBoxOfficeList.json?key=cfb1a06010c5abdf76058f51ef1816cf&targetDt={date[j[i]]}&weekGb=0"
    request = ul.Request(url)
    
    response = ul.urlopen(request)
    rescode = response.getcode()

    if(rescode == 200): responseData = response.read()
        
    result=json.loads(responseData)
    
    data.append(result)
```

```python
data
```

```python
data[0]
```

```python
movieCd=[]
for i in range(50):
    for j in range(10):
        a=data[i].get('boxOfficeResult').get('weeklyBoxOfficeList')[j].get('movieCd')
        movieCd.append(a)
```

```python
movieCd=pd.Series(movieCd)
```

```python
movieCd=pd.DataFrame(movieCd)
```

```python
movieCd
```

```python
movieNm=[]
for i in range(50):
    for j in range(10):
        a=data[i].get('boxOfficeResult').get('weeklyBoxOfficeList')[j].get('movieNm')
        movieNm.append(a)
```

```python
movieNm=pd.Series(movieNm)
```

```python
movieNm=pd.DataFrame(movieNm)
```

```python
audiAcc=[]
for i in range(50):
    for j in range(10):
        a=data[i].get('boxOfficeResult').get('weeklyBoxOfficeList')[j].get('audiAcc')
        audiAcc.append(a)
```

```python
audiAcc=pd.Series(audiAcc)
```

```python
audiAcc=pd.DataFrame(audiAcc)
```

```python
boxoffice=pd.concat([movieCd,movieNm,audiAcc], axis=1)
```

```python
boxoffice.columns=['영화 대표코드','영화명','누적관객수']
```
